```python
# setup single point config lists

key_names = [
    # "RIB2GP_PLYP_sp",
    # "B3LYP_D3BJ_def2_TZVPP_sp",
    # "B3LYP_D3BJ_ma-def2-TZVPP_sp",
    # "B97M-D4_sp",
    # "RI-DSD-BLYP_sp",
    # "BP86_sp",
    # "RI-DSD-PBEP86_2013_sp",
    # "PW6B95_sp",
    # "REVPBE_sp",
    # "REVPBE0_sp",
    # "SCANfunc_sp",
    # "wB97X-D4_sp",
    # "wB97X-D3BJ_sp",
    "r2SCAN3c",
    "B3LYP_D4",
    "PBEh3c",
    "PBEh3c_freq",
]

n_cores_list = [1, 2, 4, 8, 12, 16]


for key, func_string in zip(key_names, func_list):
    method, basisset, additional_settings = func_string.split("|")
    if not basisset:
        basisset = ""
    for n_cores in n_cores_list:
        key_plus_core = key + f"_{n_cores}cores_sp"

        loop_config_dict[key_plus_core] = copy.deepcopy(empty_loop_config)
        loop_config_dict[key_plus_core]["step_id"] = 1
        loop_config_dict[key_plus_core]["options"]["method"] = method.strip()
        loop_config_dict[key_plus_core]["options"]["basisset"] = basisset.strip()
        loop_config_dict[key_plus_core]["options"][
            "additional_settings"
        ] = additional_settings.strip()
        loop_config_dict[key_plus_core]["options"]["ram_per_core"] = 3500
        loop_config_dict[key_plus_core]["options"]["n_cores_per_calculation"] = n_cores
        loop_config_dict[key_plus_core]["options"]["walltime"] = "24:00:00"
```
